```python
import re
def remove_punc(ab):
    for cm in ab:
        if cm in string.punctuation:
            ab=ab.replace(cm,'') 
    return ab 
```

```python
anew['comments']=anew['comments'].apply(remove_punc)
anew 
```

```python
anew.dropna(inplace=True) 
#all the Null comments are removed
```

```python
ok=anew.head().iloc[0]['comments'] 
ok
```

```python
from textblob import TextBlob 
TextBlob(ok).sentiment.polarity 
```

```python
def find_polarity(text):
    try:
        return TextBlob(text).sentiment.polarity 
    except:
        return 0 
anew['polarity']=anew['comments'].apply(find_polarity)
anew
```

```python
ok1=anew['polarity'].value_counts() 
ok1
```

```python
ok2=ok1.reset_index() 
ok2 
```

```python
ok3=ok2.rename(columns={'index':'Polarity','polarity':'number_of_values'}) 
ok3
```

```python
ok3.set_index('Polarity').iloc[:20].plot(kind='bar')
#The comments with neutral polarity(0.0) are most in numbers which is followed by a lot of comments with relatively +ve polarity
#Hence as we can see , majority of comments are neutral and the rest are relatively on the positive side. 
```

```python
import emoji 
[char for char in anew['comments'][3] if char in emoji.EMOJI_DATA] 
```

```python
anda11=[char for ch in anew['comments'] for char in ch if char in emoji.EMOJI_DATA] 
anda11 
```

```python
 from collections import Counter 
a191=Counter(anda11).most_common(10) 
a191
```

```python
import pandas as pd 
a192=pd.DataFrame({
    'emoji':[a191[i][0] for i in range(10)],
    'frequency':[a191[i][1] for i in range(10)]
}) 
a192
```

```python
a192.set_index('emoji',inplace=True) 
a192 
```

```python
a192.plot(kind='bar') 
```
